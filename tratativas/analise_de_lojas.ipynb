### Importação dos dados



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

loja = pd.read_csv("../base-de-dados/loja_1.csv")
loja2 = pd.read_csv("../base-de-dados/loja_2.csv")
loja3 = pd.read_csv("../base-de-dados/loja_3.csv")
loja4 = pd.read_csv("../base-de-dados/loja_4.csv")

loja.head()
```

# 1. Análise do faturamento


```python
faturamento_loja1 = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

print("-------- Faturamento Total por Loja ---------")
print(f"O faturamento da Loja 1 é de: R$ {faturamento_loja1:,.2f}")
print(f"O faturamento da Loja 2 é de: R$ {faturamento_loja2:,.2f}")
print(f"O faturamento da Loja 3 é de: R$ {faturamento_loja3:,.2f}")
print(f"O faturamento da Loja 4 é de: R$ {faturamento_loja4:,.2f}")
print("---------------------------------------------")
```

# 2. Vendas por Categoria

```python
produtos_por_categoria_loja1 = loja.groupby('Categoria do Produto').size()

print("Quantidade de Produtos por Categoria - Loja 1:")
print(produtos_por_categoria_loja1)
```

```python
produtos_por_categoria_loja2 = loja2.groupby('Categoria do Produto').size()

print("\nQuantidade de Produtos por Categoria - Loja 2:")
print(produtos_por_categoria_loja2)
```

```python
produtos_por_categoria_loja3 = loja3.groupby('Categoria do Produto').size()

print("\nQuantidade de Produtos por Categoria - Loja 3:")
print(produtos_por_categoria_loja3)
```

```python
produtos_por_categoria_loja4 = loja4.groupby('Categoria do Produto').size()

print("\nQuantidade de Produtos por Categoria - Loja 4:")
print(produtos_por_categoria_loja4)
```

### Tratando vendas por categoria de forma combinada (Todas as lojas de uma vez)

```python
loja['Loja'] = 'Loja 1'
loja2['Loja'] = 'Loja 2'
loja3['Loja'] = 'Loja 3'
loja4['Loja'] = 'Loja 4'

dados_totais = pd.concat([loja, loja2, loja3, loja4], ignore_index=True)

produtos_por_categoria_total = dados_totais.groupby('Categoria do Produto').size()

print("\nQuantidade de Produtos por Categoria - Total (Todas as Lojas):")
print(produtos_por_categoria_total)
```

# Criação de gráficos

## 1. Gráfico de faturamento por lojas

```python
nomes_lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
faturamentos = [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]

plt.figure(figsize=(7, 5))

plt.bar(nomes_lojas, faturamentos, color='blue') 
plt.xlabel('Loja')
plt.ylabel('Faturamento Total (R$)')
plt.title('Faturamento Total por Loja')

for i, faturamento in enumerate(faturamentos):
    plt.text(i, faturamento + 1000, f'R${faturamento:,.2f}', ha='center')

plt.tight_layout()
plt.show()
```

## 2. Gráfico de quantidades de produtos vendidos por categoria

```python
if 'Loja' not in loja.columns:
    loja['Loja'] = 'Loja 1'
if 'Loja' not in loja2.columns:
    loja2['Loja'] = 'Loja 2'
if 'Loja' not in loja3.columns:
    loja3['Loja'] = 'Loja 3'
if 'Loja' not in loja4.columns:
    loja4['Loja'] = 'Loja 4'

dados_totais = pd.concat([loja, loja2, loja3, loja4], ignore_index=True)

contagem_produtos_por_categoria = dados_totais['Categoria do Produto'].value_counts()
categorias = contagem_produtos_por_categoria.index
contagens = contagem_produtos_por_categoria.values

plt.figure(figsize=(8, 6))
plt.barh(categorias, contagens, color='skyblue')
plt.xlabel('Quantidade Vendida')
plt.ylabel('Categoria do Produto')
plt.title('Quantidade Total de Produtos Vendidos por Categoria')

for i, contagem in enumerate(contagens):
    plt.text(contagem + 50, i, str(contagem), va='center')

plt.tight_layout()
plt.show()
```

## 3. Gráficos das notas de satisfação das lojas

```python
if 'Loja' not in loja.columns:
    loja['Loja'] = 'Loja 1'
if 'Loja' not in loja2.columns:
    loja2['Loja'] = 'Loja 2'
if 'Loja' not in loja3.columns:
    loja3['Loja'] = 'Loja 3'
if 'Loja' not in loja4.columns:
    loja4['Loja'] = 'Loja 4'

dados_totais = pd.concat([loja, loja2, loja3, loja4], ignore_index=True)

distribuicao_satisfacao = dados_totais['Avaliação da compra'].value_counts().sort_index()

notas = distribuicao_satisfacao.index
contagens_notas = distribuicao_satisfacao.values

plt.figure(figsize=(7, 5))
plt.bar(notas, contagens_notas, color='orange')

plt.xlabel('Nota de Avaliação')
plt.ylabel('Número de Avaliações')
plt.title('Distribuição das Notas de Avaliação da Compra (Total)')

plt.xticks([1, 2, 3, 4, 5])

for i, contagem in enumerate(contagens_notas):
    plt.text(notas[i], contagem + 50, str(contagem), ha='center')

plt.tight_layout()
plt.show()
```

## 4. Gráfico de desempenho geográfico

```python
dados_combinados = pd.concat([loja, loja2, loja3, loja4])

dados_combinados['lat'] = pd.to_numeric(dados_combinados['lat'], errors='coerce')
dados_combinados['lon'] = pd.to_numeric(dados_combinados['lon'], errors='coerce')
dados_combinados.dropna(subset=['lat', 'lon'], inplace=True)

lat_percentis = dados_combinados['lat'].quantile([0.01, 0.99])
lon_percentis = dados_combinados['lon'].quantile([0.01, 0.99])

lat_min_focado = lat_percentis.iloc[0]
lat_max_focado = lat_percentis.iloc[1]
lon_min_focado = lon_percentis.iloc[0]
lon_max_focado = lon_percentis.iloc[1]

plt.figure(figsize=(9, 8))

sns.kdeplot(x=dados_combinados['lon'], y=dados_combinados['lat'],
            fill=True,
            cmap='YlOrRd',
            levels=20,
            cbar=True,
            cbar_kws={'label': 'Densidade de Vendas'})

plt.xlim(lon_min_focado, lon_max_focado)
plt.ylim(lat_min_focado, lat_max_focado)

plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Mapa de Calor da Densidade de Vendas (Todas as Lojas)', fontsize=16)

plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

# Relatório de Análise das Lojas para Recomendação de Venda

## 1. Introdução

Com base nas bases de dados de vendas fornecidas, este estudo realiza uma análise detalhada de quatro lojas, explorando métricas de desempenho e características operacionais. O propósito central deste relatório é subsidiar a decisão do Senhor João, recomendando qual loja deve ser considerada para venda.

## 2. Análise Detalhada e Resultados

Nesta seção, são apresentados os resultados das análises realizadas para cada uma das métricas consideradas. Para cada análise, o código utilizado e os resultados obtidos são exibidos para transparência e reprodutibilidade.

### 2.1. Faturamento Total das Lojas

Nesta análise, calculamos a receita total gerada por cada loja somando os valores da coluna "Preço". O faturamento é uma métrica fundamental para entender o desempenho financeiro bruto de cada estabelecimento.

Com base nos valores e no gráfico, podemos observar o desempenho de cada loja em termos de receita total e percebemos que a faixa de faturamento das 4 lojas vai de 1.38M a 1.53M. A diferença total entre a maior e a menor é relativamente pequena comparada ao volume de faturamento. Não tem um faturamento idêntico, mas também não tem uma loja que fique muito a frente ou muito atrás das outras.

### 2.2. Categorias de Produtos Mais e Menos Vendidas

Analisamos a distribuição de vendas por categoria de produto para entender o mix de produtos que mais contribui para o volume de vendas. Inicialmente, focamos na contagem total por categoria considerando todas as lojas.

O gráfico de barras horizontal ilustra claramente a distribuição do volume de vendas entre as diferentes categorias na base de dados total. Podemos observar que Móveis e Eletrônicos são as categorias de produto com o maior volume de vendas quando consideramos todas as lojas juntas. Esta análise geral nos dá uma visão do mix de produtos mais popular em toda a operação.

### 2.3. Média das Avaliações de Clientes por Loja

A satisfação do cliente é um indicador crucial da qualidade do serviço e dos produtos. Calculamos a média das avaliações de compra (em uma escala de 1 a 5) para cada loja para entender a percepção geral dos clientes.

Os resultados indicam as médias de satisfação para cada loja são bem positivas, a grande maioria das avaliações são nota 5.

### 2.4. Produtos Mais e Menos Vendidos por Loja

Identificamos o produto individual mais vendido e o produto com a menor contagem de vendas em cada loja.

Analisando os extremos, cada loja possui produtos distintos como os mais e menos vendidos. **No entanto, o volume de vendas tanto dos produtos mais vendidos (entre 57 e 65 unidades) quanto dos menos vendidos (entre 32 e 35 unidades) é bastante similar entre as lojas.** Isso sugere um comportamento de vendas comparável nos extremos de performance de produtos.

### 2.5. Frete Médio de Cada Loja

Calculamos o valor médio do frete associado às vendas de cada loja.

A Loja 1 apresentou o maior frete médio (R$ 34,69), enquanto a Loja 4 teve o menor (R$ 31,28). **Observamos uma correlação aparente onde a loja com maior faturamento (Loja 1) teve o maior frete médio, e a loja com menor faturamento (Loja 4) teve o menor frete médio.** A diferença absoluta no frete médio entre as lojas não é muito grande.

### 2.6. Análise Geográfica

Exploramos os dados de latitude e longitude para visualizar a distribuição geográfica das vendas.

Os gráficos geográficos e a análise das coordenadas mais frequentes revelam que as **vendas estão fortemente concentradas em poucos pontos geográficos exatos**. Isso sugere que estes pontos correspondem às localizações físicas das lojas ou hubs de venda, em vez de endereços de clientes dispersos. A análise mostra onde a densidade de vendas é mais alta na área de atuação principal.

## 3. Conclusão e Recomendação

Com base nas análises detalhadas apresentadas, podemos comparar o desempenho e as características de cada loja em relação ao faturamento, mix de produtos (categorias/produtos mais vendidos), satisfação do cliente, custo de frete e distribuição geográfica das vendas.

**Resumo Comparativo:**

* **Loja 1:** Pontos fortes: Maior faturamento, Pontos fracos: Maior taxa de frete, Menor média de satisfação dos clientes
* **Loja 2:** Pontos fortes: É a segunda loja com maior faturamento, Pontos fracos: Embora tenha um produto mais vendido notável dentro da loja (iniciando em programação com 65 unidades), o volume desse produto é relativamente baixo quando comparado aos produtos de maior venda na base de dados total.
* **Loja 3:** Pontos fortes: Maior média de satisfação dos clientes, Pontos fracos: É a segunda loja com o menor faturamento
* **Loja 4:** Pontos fortes: É a loja com o menor taxa de frete, Pontos fracos: É a loja com o menor faturamento, Segunda menor média de satisfação dos clientes

Considerando todos esses fatores, a recomendação para o Senhor João sobre qual loja vender é a **Loja 4**.

**Justificativa da Recomendação:**

A decisão de recomendar a venda da Loja 4 é baseada nos seguintes pontos principais observados na análise:

1.  **Menor Faturamento Total:** É a loja com o **menor faturamento total** (R$ 1,384,497.58), conforme observado na Seção 2.1 e ilustrado no gráfico de faturamento. Este é o indicador financeiro mais direto e a posiciona como a mais fraca em termos de receita bruta.
2.  **Baixa Satisfação do Cliente (Segunda Menor):** A Loja 4 registrou a **segunda menor média de avaliação dos clientes** (4.0 em uma escala de 1 a 5), conforme detalhado na Seção 2.3. Embora a Loja 1 tenha a menor média absoluta de satisfação, a satisfação na Loja 4 ainda é relativamente baixa e a coloca entre as lojas com menor índice de satisfação, contribuindo para seus pontos fracos gerais.
3.  **Desafios Geográficos Potenciais:** As análises geográficas (Seção 2.6) sugerem que embora suas vendas estejam bem distribuidas podem estar em uma área com menor densidade geral de vendas.

Apesar de possuir o menor frete médio, o impacto combinado do **menor faturamento total** e da **baixa satisfação do cliente** a torna a candidata mais fraca para ser mantida no portfólio. Enquanto a Loja 1 tem o maior faturamento apesar da menor satisfação, e a Loja 2 tem o segundo maior faturamento mesmo com bestsellers de volume limitado no contexto geral, a Loja 4 apresenta o pior desempenho financeiro e uma baixa satisfação (sendo a segunda pior), o que, em conjunto, a torna a mais indicada para ser vendida em comparação com as outras.

Em suma, a Loja 4 demonstra ser a loja menos promissora dentre as 4, apresentando os resultados mais fracos em faturamento e baixa satisfação do cliente (a segunda menor), além de potenciais desafios geográficos que a posicionam como a mais indicada para ser vendida.
